```python
books = pd.read_csv('book.csv')
```

### Aripori Algorithm

### 1. Association rules with 10% Support and 100% confidence

```python
#10% support
frequent_itemsets = apriori(books, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
#100% confidence
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.sort_values('lift',ascending = False).head(10)
```

```python
r1 = rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules[rules.lift>1]
```

```python

import matplotlib.pyplot as plt
plt.scatter(rules['support'], rules['confidence'], alpha=0.5)
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('Support vs Confidence')
plt.show()
```

```python
plt.scatter(rules['support'], rules['confidence'], alpha=0.5)
plt.xlabel('support')
plt.ylabel('lift')
plt.title('Support vs Liftt')
plt.show()
```

```python
import numpy as np
fit = np.polyfit(rules['lift'], rules['confidence'], 1)
fit_fn = np.poly1d(fit)
plt.plot(rules['lift'], rules['confidence'], 'yo', rules['lift'],
 fit_fn(rules['lift']))
```

```python
rules[rules.lift>1].head(20)
```

```python
for index, row in r1.iterrows():
  val1 = row['antecedents']
  val2 = row['consequents']
  conf = row['confidence']
  sup = row['support']
  lift = row['lift']
  print("Rule"+str(index)+": ",val1,"==>",val2)
  print("confidence : ",conf)
  print("support : ",sup)
  print("lift : ",lift)
  print("\n")
```
